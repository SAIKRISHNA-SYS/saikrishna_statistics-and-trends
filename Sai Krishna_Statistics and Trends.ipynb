```python
#IMPORTING ALL THE REQUIRED LIBRARIES
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Downloads\\mental_health_dataset.csv") # Reading data from csv file
data = df.sample(n=100) # storing the first hundred values of dataset in an variable named data
```

```python
#cleaning data and pre-processing of data
data.dropna()
data.drop_duplicates()
```

```python
statistics_description = data.describe()
print(statistics_description)
```

```python
    plt.figure(figsize=(10, 6)) # Defining Figure Size
    sns.histplot(data=data,x="Work_Hours",hue="Gender", multiple="stack",kde=True) # plotting histogram for type of work
    plt.xlabel("Work Hours") #Setting X Label
    plt.ylabel("Frequency") # Setting Y label
    plt.show() # showing the plot
```

```python
    plt.figure(figsize=(10, 6)) # Defining Figure Size
    sns.scatterplot(data=data, x="Age", y="Work_Hours",hue ="Gender") # Plotting scatter plot
    plt.title("Age vs.Work Hours") # Setting Plot Title
    plt.xlabel("Age") # Setting X label
    plt.ylabel("Work Hours") # Setting Y label
    plt.show() # showing plot
```

```python
# plotting of heatmap to understand the correaltion between the variables of data set
numeric_data = data.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='crest')
plt.title("Correlation Matrix of Numeric Features")
plt.show()
```
